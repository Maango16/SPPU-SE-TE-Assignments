```python
page = requests.get(link)

page
```

```python
soup = bs(page.content,'html.parser')

print(soup.prettify())
```

```python
names = soup.find_all('p',class_='_2sc7ZR _2V5EHH')

names
```

```python
cust_name = []

for i in range(0,len(names)):

    cust_name.append(names[i].get_text())

cust_name
```

```python
title = soup.find_all('p',class_='_2-N8zT')

title
```

```python
buyer_type = soup.find_all('p',class_='_2mcZGG')

buyer_type
```

```python
cust_add = []

for i in range(0,len(buyer_type)):

    cust_add.append(buyer_type[i].get_text())

cust_add
```

```python
df = pd.DataFrame()
```

```python
df['Customer Name']=cust_name
```

```python
df['Review title']=title
df['Customer Address']=cust_add
```
